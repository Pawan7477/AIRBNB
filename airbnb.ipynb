```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# handling missing value

```python
df['last review'] =pd.to_datetime(df['last review'], errors ='coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0, 'last review' : df['last review'].min()}, inplace =True)
```

```python
df.dropna(subset=['NAME', 'host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df= df.drop(columns=["license", "house_rules"], errors='ignore')
```

```python
df.head()
```

```python
# Remove dollar sign and commas, then convert to float
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
df['service fee'] = df['service fee'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
df.head()
```

# remove duplicate value

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# visualization

# what is the distribution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde =True, color='red')
plt.title("Distribution of listing price")
plt.xlabel("price $")
plt.ylabel("Frequency")
plt.show()
```

# how are different room types distribution ?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot( x= "room type", data = df,color="hotpink")
plt.title("Room type distribution")
plt.xlabel("room type")
plt.ylabel("count")
plt.show()
```

# How are listings distributedd across different neighborhoods?

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='neighbourhood group', data=df, color="lightgreen", order=df['neighbourhood group'].value_counts().index)
plt.title('Number of Neighborhood Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

# what is the relationship between price and room type

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the boxplot with the 'Set1' color palette
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')

# Set the title and labels
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')

# Display the legend with the title 'Room Type'
plt.legend(title='Room Type')

# Show the plot
plt.show()
```

# how number of reviews change over with time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the 'last review' column to datetime
df['last review'] = pd.to_datetime(df['last review'])

# Group by month and count reviews over time
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

# Plot the number of reviews over time
plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
